```python
!pip install google-generativeai pillow -q

import google.generativeai as genai
from PIL import Image as PILImage
from google.colab import files
import io

# Configure API
genai.configure(api_key="AIzaSyA2PLCZHtUmrx4sVTDYhz89qpG5yyqp7w8")

print("=" * 80)
print(" MEDICAL IMAGING ANALYSIS - Colab Version")
print("=" * 80)
print("\n Click 'Choose Files' to upload your medical image\n")

# Upload image
uploaded = files.upload()

if uploaded:
    for filename in uploaded:
        print(f"\n{'='*80}")
        print(f"‚úÖ Image uploaded: {filename}")
        print(f"{'='*80}\n")

        # Open image
        image = PILImage.open(io.BytesIO(uploaded[filename]))

        # Display image info
        print(f"Image size: {image.size}")
        print(f"Image format: {image.format}\n")

        # Show the image
        image.show()

        print("\n ANALYZING IMAGE... (This takes 30-60 seconds)\n")
        print("-" * 80)

        # Initialize model
        model = genai.GenerativeModel('gemini-2.0-flash')

        # Analysis prompt
        analysis_prompt = """You are a highly skilled medical imaging expert. Analyze this medical image and provide:

### 1. Image Type & Region
- Imaging modality (X-ray/MRI/CT/Ultrasound/etc.)
- Anatomical region
- Image quality assessment

### 2. Key Findings
- Primary observations
- Abnormalities (if any)
- Measurements where relevant
- Severity rating (Normal/Mild/Moderate/Severe)

### 3. Diagnostic Assessment
- Primary diagnosis with confidence level
- Differential diagnoses
- Supporting evidence

### 4. Patient-Friendly Explanation
- Simple language explanation
- Visual analogies
- Common concerns addressed

### 5. Additional Context
- Relevant medical information
- Suggested next steps
- Key references

Format clearly with headers and bullet points."""

        try:
            # Run analysis
            response = model.generate_content([
                analysis_prompt,
                image
            ])

            print("\nüìã ANALYSIS RESULTS:")
            print("-" * 80)
            print(response.text)
            print("-" * 80)
            print("\n‚ö†Ô∏è  DISCLAIMER: This analysis is AI-generated.")
            print("Always consult qualified healthcare professionals for medical decisions.")

        except Exception as e:
            print(f"\n‚ùå Error during analysis: {str(e)}")
            print("Please check your API key and try again.")
else:
    print("‚ùå No image uploaded. Please try again.")
```
